```python
#steering
MAX_STEER_DEGREES = 40

def angle_between(v1, v2):
    return math.degrees(np.arctan2(v1[1], v1[0]) - np.arctan2(v2[1], v2[0]))

# function to get angle between the car and target waypoint
def get_angle(car,wp):
    '''
    this function to find direction to selected waypoint
    '''
    vehicle_pos = car.get_transform()
    car_x = vehicle_pos.location.x
    car_y = vehicle_pos.location.y
    wp_x = wp.transform.location.x
    wp_y = wp.transform.location.y
    
    # vector to waypoint
    x = (wp_x - car_x)/((wp_y - car_y)**2 + (wp_x - car_x)**2)**0.5
    y = (wp_y - car_y)/((wp_y - car_y)**2 + (wp_x - car_x)**2)**0.5
    
    #car vector
    car_vector = vehicle_pos.get_forward_vector()
    degrees = angle_between((x,y),(car_vector.x,car_vector.y))

    return degrees
```

```python
# Function to set throttle and brake based on current speed and target speed
def control_speed(vehicle, target_speed, throttle, brake,steer_input):
    current_speed = 3.6 * vehicle.get_velocity().length()  # m/s to km/h
    #throttle=target_speed/30 *1
    if current_speed < target_speed:
        vehicle.apply_control(carla.VehicleControl(throttle=throttle, brake=0.0,steer=steer_input))
    else:
        vehicle.apply_control(carla.VehicleControl(throttle=0.0, brake=brake,steer=steer_input))

def calculating_acceleration_speed(currentPoint, accelerationPoints):
    maxSpeed=30
    minSpeed =10
    accelerationFactor = 1.3
    
    # Calculate the waypoint index at which acceleration should stop
    accelerationEndPoint = round(accelerationPoints * 0.25)
    
    if (currentPoint <= accelerationEndPoint):
        # Quadratic acceleration function
        speed = min(minSpeed + accelerationFactor * ((currentPoint / accelerationPoints), maxSpeed)**.5)
    else:
        # Maintain max speed for the rest of the waypoints
        speed = maxSpeed
    return speed


def calculateCoastingSpeed(currentPoint, coastingPoints, accelerationPoints):
    
    #Adjust these parameters as needed
    maxSpeed = 30
    minSpeed = 10 
    decelerationFactor = 1.3
    
    coastingStartPoint = accelerationPoints + 1
    coastingEndPoint = accelerationPoints + coastingPoints
    
    #Calculate the waypoint index at which deceleration should start
    decelerationStartPoint = coastingStartPoint + round(coastingPoints * 0.25)

    if currentPoint >= decelerationStartPoint and  currentPoint <= coastingEndPoint:
        #Quadratic deceleration during coasting
        relativeCoastingPoint = currentPoint - coastingStartPoint + 1
        speed = max(minSpeed, maxSpeed - decelerationFactor * ((relativeCoastingPoint / coastingPoints)**0.5))
    else:
        # Maintain min speed for the first 25% of coasting waypoints
        speed = minSpeed
    return speed
```
